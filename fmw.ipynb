```python
from word_generator import WordGenerator
from pronunciation_generator import PronunciationGenerator
from meaning_generator import MeaningGenerator
from image_generator import ImageGenerator

from datetime import datetime

from IPython.display import display
```

```python
# Insert your openai_api_key
openai_api_key = ""
```

```python
# Initialize the generators
t = datetime.now()

wg = WordGenerator()
pg = PronunciationGenerator(openai_api_key)
mg = MeaningGenerator()
ig = ImageGenerator()

!mkdir image_samples

print("Loading took", datetime.now()-t)
```

```python
# Generate a fake word
print("Generating new fake word...")
t = datetime.now()
word, pos, _, affix_meanings = wg.generate()
pronunciation = pg.generate(word)
meaning, example = mg.generate(word, pos, affix_meanings)
image = ig.generate(word, pos, pronunciation, meaning)

print(f"{word}, {pos}, {pronunciation}")
print(f"Definition: {meaning}")
print(f"Example sentence: {example}")
display(image)
image.save(f"image_samples/{word}.png")
print("Generation took", datetime.now()-t)
```
